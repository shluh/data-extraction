# Extraindo dados de um arquivo CSV

## Importando pacotes necessários para ler arquivos CSV

```python
import pandas as pd
import csv
```

# Lendo o arquivo CSV e exibindo os resultados

```python
with open('part-00000-3910dd5d-b928-43b9-a3cc-8b47b3233936.c000.csv') as csv_file:
    csv_reader = csv.reader(csv_file, delimiter=';')
    line_count = 0
    for row in csv_reader:
        if line_count < 15:
            if line_count == 0:
                print(f'Column names are {", ".join(row)}')
                line_count += 1
            else:
                print(f'\t{row[0]} works in the {row[1]} department, and was born in {row[2]}.')
                line_count += 1
    print(f'Processed {line_count} lines.')

```

## Lendo o arquivo CSV usando pandas e exibindo

```python
data= pd.read_csv("part-00000-3910dd5d-b928-43b9-a3cc-8b47b3233936.c000.csv", sep=";")
data
```

# Extraindo Dados de uma pagina Web

## Importando pacotes para realizar a extração e tratamentos dos dados da página

```python
import requests
from bs4 import BeautifulSoup
```

## Realizando extração do HTML da página

```python
page = requests.get("https://quotes.toscrape.com/")
```

## Verficando o conteúdo da página extraída

```python
page.content
```

##  Convertendo o resultado da extração para um HTML

```python
soup = BeautifulSoup(page.content, 'html.parser')
```

## Exibindo resultado da extração e conversão de forma a ser lida

```python
print(soup.prettify())
```

## Buscando conteúdo dado tag e suas caracteristicas

```python
soup.find_all('p')[0].get_text()
```

## Buscando tag dando um padrão

```python
soup.select("div p")
```
